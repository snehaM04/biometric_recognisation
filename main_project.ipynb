# Load the dataset

```python
import os
import cv2
import numpy as np

def load_images_and_labels(base_dir, img_size=(224, 224)):
    images = []
    labels = []
    
    # Loop over each folder in the dataset directory
    for folder_name in os.listdir(base_dir):
        folder_path = os.path.join(base_dir, folder_name)
        
        # Ensure that we are only processing directories (i.e., label folders)
        if os.path.isdir(folder_path):
            # Loop over each image in the folder
            for image_name in os.listdir(folder_path):
                img_path = os.path.join(folder_path, image_name)
                print(f"Loading image: {img_path}") 
                # Read and resize the image
                img = cv2.imread(img_path)  # Load the image in color (BGR format)
                #print(img.shape)
                if img is None:
                    print(f"Failed to load image: {img_path}")
                    continue
                
                img = cv2.resize(img, img_size)  # Resize to match EfficientNet input size
                
                # Append the image and the label (folder name)
                images.append(img)
                labels.append(int(folder_name))  # Assuming folder name is the label (e.g., 1, 2, 3, etc.)
    
    # Convert lists to NumPy arrays
    images = np.array(images)
    labels = np.array(labels)
    
    print("Before normalizing the images:")
    print("Images shape:", images.shape)
    print("Images:", images)
    print("Images type:",images.dtype)
    
    # Normalize images to the range [0, 1]
    images = images.astype(np.float32) / 255.0
    print("After normalizing:")
    print("Images shape:", images.shape)
    print("Images:", images)
    print("Images type",images.dtype)
    
    # Check if the images are grayscale or RGB
    if len(images.shape) == 3 and images.shape[2] == 1:  # Single-channel grayscale
        print("The images are Grayscale")
    elif len(images.shape) == 4 and images.shape[3] == 3:  # Three-channel RGB
        print("The images are RGB")
    else:
        print("Unknown image format")
  
    print(f"Loaded {len(images)} images after quality check.")
    print("Labels:", labels)
    print("Labels shape:", labels.shape)
   

    return images, labels

# Load images and labels
image_dir = "C:/iris_dataset"  # Replace with the correct path

images, labels = load_images_and_labels(image_dir) 
```

# Preprocessing the dataset using Wiener filtering

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import wiener
import os

# Function to apply Wiener filtering to grayscale images
def wiener_filter_grayscale(image):
    # Apply Wiener filter to the grayscale image with a kernel size of 3x3
    filtered_image = wiener(image, (3, 3))
    return filtered_image

# Function to apply Wiener filtering to RGB images (converted to grayscale first)
def wiener_filter_rgb_to_grayscale(image):
    # Convert RGB image to grayscale
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Apply Wiener filter to the grayscale image with a kernel size of 3x3
    filtered_image = wiener(grayscale_image, (3, 3))
    return grayscale_image, filtered_image

# Function to apply Wiener filtering on a list of images and segment them
def preprocess_and_segment_images(images):
    processed_images = []
    segmented_images = []
    iris_masks = []
    
    for idx, image in enumerate(images):
        print(f"Processing and segmenting image {idx+1}/{len(images)}")
        print(image.shape)

         # Process as grayscale or RGB
        if len(images.shape) == 3 and images.shape[2] == 1:  # Grayscale image
            print(f"Processing grayscale image {idx+1}")
            wiener_filtered_image = wiener_filter_grayscale(image)

        elif len(images.shape) == 4 and images.shape[3] == 3:   # RGB image
            print(f"Processing RGB image {idx+1}")
            _, wiener_filtered_image = wiener_filter_rgb_to_grayscale(image)

        else:
            print(f"Unsupported image format for image {idx+1}. Skipping.")
            continue
        
        

        # Perform iris segmentation on the processed image
        iris_mask, segmented_iris = iris_segmentation_debug(wiener_filtered_image)
        

        if iris_mask is not None and segmented_iris is not None:
            # Store the segmented image and mask
            segmented_images.append(segmented_iris)
            iris_masks.append(iris_mask)

            # Visualize only the first segmented image
            if idx == 0:
                visualize_images(image, wiener_filtered_image, is_rgb=(len(image.shape) == 4 and image.shape[3] == 3))
                visualize_segmentation(image, iris_mask, segmented_iris)
                print("Original iris image:", np.array(image))
                print("Original iris image shape:", image.shape)
                print("Filtered iris image shape:", wiener_filtered_image.shape)
                print("Filtered iris image:", wiener_filtered_image)
                print("segmented iris shape:", len(segmented_images))                         

        # Append the processed (filtered) image
        processed_images.append(wiener_filtered_image)

    print(f"Processing and segmentation completed for {len(processed_images)} images.")
    return np.array(processed_images), np.array(segmented_images), np.array(iris_masks)

def iris_segmentation_debug1(image):
    print("Image format:", image.dtype)
    
    # Convert image to 8-bit (uint8) format
    image_uint8 = np.uint8(255 * (image - np.min(image)) / (np.max(image) - np.min(image)))

    # Adaptive Histogram Equalization (CLAHE)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    image_eq = clahe.apply(image_uint8)

    # Gaussian blur to reduce noise
    blurred_image = cv2.GaussianBlur(image_eq, (9, 9), 2)

    # Hough Circle Transform to detect the iris
    # Adjust minRadius and maxRadius based on image size (380x380)
    min_radius = 30  # Adjust this value based on the expected iris size
    max_radius = 150  # Adjust this value based on the expected iris size

    circles = cv2.HoughCircles(blurred_image, cv2.HOUGH_GRADIENT, dp=1.2, minDist=100,
                               param1=100, param2=30, minRadius=min_radius, maxRadius=max_radius)
    print("Using Hough Circle Transform to detect the iris:", circles)

    if circles is not None:
        circles = np.round(circles[0, :]).astype("int")
        x, y, r = circles[0]

        # Draw detected circle for verification
        output_image = image_uint8.copy()
        cv2.circle(output_image, (x, y), r, 255, -1)

        # Create a mask for the iris
        mask = np.zeros_like(image_uint8)
        cv2.circle(mask, (x, y), r, 255, -1)

        # Post-process mask to refine boundaries (adjusted kernel size)
        kernel = np.ones((3, 3), np.uint8)  # Adjust kernel size
        mask_eroded = cv2.erode(mask, kernel, iterations=1)

        # Segment the iris using the mask
        segmented_iris = cv2.bitwise_and(image_uint8, image_uint8, mask=mask_eroded)

        # Normalize the segmented iris for visualization
        segmented_iris_normalized = cv2.normalize(segmented_iris, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)

        return mask_eroded, segmented_iris_normalized
    else:
        print("Iris not detected.")
        return None, None


def iris_segmentation_debug(image):
    print("Image format:", image.dtype)
    
    # Convert image to 8-bit (uint8) format
    image_uint8 = np.uint8(255 * (image - np.min(image)) / (np.max(image) - np.min(image)))
    #print("Converted image to 8-bit (uint8):", image_uint8.dtype)

    # Adaptive Histogram Equalization (CLAHE)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    image_eq = clahe.apply(image_uint8)

    # Gaussian blur to reduce noise
    blurred_image = cv2.GaussianBlur(image_eq, (9, 9), 2)


    # Hough Circle Transform to detect the iris
    circles = cv2.HoughCircles(blurred_image, cv2.HOUGH_GRADIENT, dp=1.0, minDist=50, param1=150, param2=20, minRadius=20, maxRadius=100)
    print("Using Hough Circle Transform to detect the iris:", circles)
    
    if circles is not None:
        circles = np.round(circles[0, :]).astype("int")
        x, y, r = circles[0]

        # Draw detected circle for verification
        output_image = image_uint8.copy()
        #print(output_image)
        cv2.circle(output_image, (x, y), r,255, -1)

        # Create a mask for the iris
        mask = np.zeros_like(image_uint8)
        cv2.circle(mask, (x, y), r, 255, -1)

        # Post-process mask to refine boundaries (adjusted kernel size)
        kernel = np.ones((3, 3), np.uint8)  # Reduced kernel size to avoid over-erosion
        mask_eroded = cv2.erode(mask, kernel, iterations=1)

        # Segment the iris using the mask
        segmented_iris = cv2.bitwise_and(image_uint8, image_uint8, mask=mask_eroded)
        
        # Normalize the segmented iris for visualization (fixed normalization step)
        segmented_iris_normalized = cv2.normalize(segmented_iris, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
      
        return mask_eroded, segmented_iris_normalized
    else:
        print("Iris not detected.")
        return None, None




# Function to visualize one original and filtered image
def visualize_images(original_image, filtered_image, is_rgb=False):
    plt.figure(figsize=(10, 5))
    
    # Show original image
    plt.subplot(1, 2, 1)
    if is_rgb:
        plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
    else:
        plt.imshow(original_image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # Show filtered image
    plt.subplot(1, 2, 2)
    plt.imshow(filtered_image, cmap='gray')
    plt.title('Wiener Filtered Image (3x3 Kernel)')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
    
    
# Function to visualize the segmented iris and mask
def visualize_segmentation(original_image, iris_mask, segmented_iris):
    plt.figure(figsize=(10, 5))

    # Show original image
    plt.subplot(1, 2, 1)
    plt.imshow(original_image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # Show segmented iris mask
    plt.subplot(1, 2, 2)
    plt.imshow(iris_mask, cmap='gray')
    plt.title('Iris Mask')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

# Example usage (assuming `images` is defined):
processed_images, segmented_images, iris_masks = preprocess_and_segment_images(images)

```

# Feature Extraction using EfficientNetB4 model 

```python
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Model
import numpy as np
import cv2

# Function to convert grayscale to 3-channel image
def grayscale_to_rgb(image):
    return cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)

# EfficientNetB7 model input size
effnet_input_size = 224

# Custom Swish activation function
def swish(x):
    return x * tf.keras.activations.sigmoid(x)

# Function to load EfficientNetB4 model with Swish activation and preserve spatial features
def load_efficientnetb0_with_swish():
    input_size = effnet_input_size
    
    # Load pre-trained EfficientNetB4 model without the top layer
    base_model = EfficientNetB0(input_shape=(input_size, input_size, 3), include_top=False, weights='imagenet')
    
    # Modify the base model to use Swish activation
    for layer in base_model.layers:
        if hasattr(layer, 'activation'):
            if layer.activation == tf.keras.activations.relu:
                layer.activation = swish
    
    # Return the model up to the convolutional layers (without pooling)
    model = Model(inputs=base_model.input, outputs=base_model.output)
    
    return model, input_size

# Function to preprocess the image and extract features using EfficientNetB4
def extract_features_efficientnetb0(segmented_images):
    # Load the EfficientNetB4 model with Swish activation
    model, input_size = load_efficientnetb0_with_swish()
    
    # Initialize a list to store the extracted features
    feature_list = []
    
    # Iterate through each segmented image
    for idx, image in enumerate(segmented_images):
        print(f"Extracting features from image {idx+1}/{len(segmented_images)} using EfficientNetB0...")
        
        # Resize the image to the required input size of EfficientNetB7
        resized_image = cv2.resize(image, (input_size, input_size))
        
        # Convert the grayscale image to 3 channels (RGB)
        rgb_image = grayscale_to_rgb(resized_image)
        
        # Normalize the image (to the range [0, 1])
        rgb_image = rgb_image.astype('float32') / 255.0
        
        # Add batch dimension
        rgb_image = np.expand_dims(rgb_image, axis=0)
        
        # Extract features using the model
        features = model.predict(rgb_image)
        
        # Squeeze to remove any singleton dimensions (like the extra 1)
        features = np.squeeze(features)
        
        # Append the extracted features to the list
        feature_list.append(features)
        
        
    # Convert the list of features to a NumPy array and reshape if necessary
    return np.array(feature_list)

# Example usage with the segmented iris images
# Assuming `segmented_images` are the output from your segmentation step (shape: (1010, 224, 224))

# Resize the images and extract features using EfficientNetB7
features = extract_features_efficientnetb0(segmented_images)

# Print the shape of the extracted features
print(f"Features extracted using EfficientNetB0: {features.shape}")

print("Feature extraction completed.")
```

# Hyperparameter tuning using Orca Predator Algorithm

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Model
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.optimizers import Adam
from sklearn.datasets import make_classification

# OPA (Orca Predator Algorithm) for Feature Optimization
class OrcaPredatorAlgorithm:
    def __init__(self, n_predators, n_features, max_iter, features, labels):
        self.n_predators = n_predators
        self.n_features = n_features
        self.max_iter = max_iter
        self.features = features
        self.labels = labels
        self.position = np.random.randint(2, size=(n_predators, n_features)).astype(float)  # Convert to float after initialization
        self.velocity = np.random.uniform(-1, 1, size=(n_predators, n_features))  # Velocity
        self.best_position = np.copy(self.position)  # Best position
        self.best_fitness = np.full(n_predators, -np.inf)  # Best fitness for each predator
        self.global_best_position = np.zeros(n_features)  # Global best position
        self.global_best_fitness = -np.inf  # Global best fitness

    # Function to evaluate fitness based on feature selection
    def evaluate_fitness(self, predator_position):
        selected_features = np.where(predator_position == 1)[0]  # Get selected feature indices
        
        # If no features are selected, return a low fitness value
        if len(selected_features) == 0:
            return -np.inf

        # Use only the selected features
        X_selected = self.features[:, selected_features]
        
        # Split the data into train and test sets
        X_train, X_test, y_train, y_test = train_test_split(X_selected, self.labels, test_size=0.3, random_state=42)

        # Define a simple neural network model for classification
        model = self.build_classifier(X_train.shape[1], n_classes=len(np.unique(self.labels)))  # Modify for multi-class
        
        # Train the model
        model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=0)
        
        # Evaluate on the test set
        y_pred = model.predict(X_test)
        y_pred_classes = np.argmax(y_pred, axis=1)  # Multi-class prediction using argmax
        accuracy = accuracy_score(y_test, y_pred_classes)
        
        return accuracy

    # Function to build a simple classifier model
    def build_classifier(self, input_shape, n_classes):
        inputs = Input(shape=(input_shape,))
        x = Dense(64, activation='relu')(inputs)
        x = Dense(32, activation='relu')(x)
        output = Dense(n_classes, activation='softmax')(x)  # Multi-class classification using softmax
        
        model = Model(inputs=inputs, outputs=output)
        model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])  # Sparse categorical cross-entropy for multi-class
        return model
    
    # Orca movement behavior
    def move_predators(self):
        for i in range(self.n_predators):
            # Update velocity based on best predator
            self.velocity[i] = 0.5 * self.velocity[i] + np.random.uniform(0, 1) * (self.best_position[i] - self.position[i]) + np.random.uniform(0, 1) * (self.global_best_position - self.position[i])
            
            # Update position
            self.position[i] += self.velocity[i]
            
            # Ensure the position remains binary (0 or 1)
            self.position[i] = np.clip(self.position[i], 0, 1)
        
    # Update best position and fitness
    def update_best_position(self):
        for i in range(self.n_predators):
            fitness = self.evaluate_fitness(self.position[i])
            if fitness > self.best_fitness[i]:
                self.best_fitness[i] = fitness
                self.best_position[i] = self.position[i]
            
            # Update global best
            if fitness > self.global_best_fitness:
                self.global_best_fitness = fitness
                self.global_best_position = self.position[i]
    
    # Run the OPA algorithm
    def run(self):
        for iteration in range(self.max_iter):
            print(f"Iteration: {iteration + 1}/{self.max_iter}")
            self.update_best_position()
            self.move_predators()

        return self.global_best_position, self.global_best_fitness

# Run OPA to optimize the feature set
n_predators = 10  # Number of predators (agents)
max_iter = 20     # Number of iterations

# Initialize and run the OPA algorithm
opa = OrcaPredatorAlgorithm(n_predators=n_predators, n_features=scaled_features.shape[1], 
                            max_iter=max_iter, features=scaled_features, labels=labels)

# Run the optimization
best_features, best_fitness = opa.run()

print(f"Best Fitness: {best_fitness}")
print(f"Best Feature Set: {best_features}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, Flatten, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.datasets import make_classification

# Convolutional Autoencoder Definition
class ConvolutionalAutoencoder:
    def __init__(self, input_shape, n_classes):
        self.input_shape = input_shape
        self.n_classes = n_classes
        self.autoencoder = self.build_autoencoder()
        self.classifier = self.build_classifier()

    def build_autoencoder(self):
        # Encoder
        input_img = Input(shape=self.input_shape)
        x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
        x = MaxPooling2D((2, 2), padding='same')(x)
        x = Conv2D(16, (3, 3), activation='relu', padding='same')(x)
        x = MaxPooling2D((2, 2), padding='same')(x)
        x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
        encoded = MaxPooling2D((2, 2), padding='same')(x)

        # Decoder
        x = Conv2D(8, (3, 3), activation='relu', padding='same')(encoded)
        x = UpSampling2D((2, 2))(x)
        x = Conv2D(16, (3, 3), activation='relu', padding='same')(x)
        x = UpSampling2D((2, 2))(x)
        x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
        x = UpSampling2D((2, 2))(x)
        decoded = Conv2D(self.input_shape[2], (3, 3), activation='sigmoid', padding='same')(x)

        autoencoder = Model(input_img, decoded)
        autoencoder.compile(optimizer=Adam(), loss='binary_crossentropy')
        return autoencoder

    def build_classifier(self):
        # Define the classifier based on the encoder part of the autoencoder
        input_img = Input(shape=self.input_shape)
        x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
        x = MaxPooling2D((2, 2), padding='same')(x)
        x = Conv2D(16, (3, 3), activation='relu', padding='same')(x)
        x = MaxPooling2D((2, 2), padding='same')(x)
        x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
        x = MaxPooling2D((2, 2), padding='same')(x)
        x = Flatten()(x)  # Flatten for classification
        x = Dense(128, activation='relu')(x)  # Fully connected layer
        output = Dense(self.n_classes, activation='softmax')(x)  # Output layer for classification

        classifier = Model(input_img, output)
        classifier.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        return classifier

    def train_autoencoder(self, X, epochs=50, batch_size=256):
        self.autoencoder.fit(X, X, epochs=epochs, batch_size=batch_size, shuffle=True, validation_split=0.2)

    def train_classifier(self, X, y, epochs=50, batch_size=256):
        self.classifier.fit(X, y, epochs=epochs, batch_size=batch_size, validation_split=0.2)

# Generate synthetic data for demonstration
n_samples = 2971
img_height, img_width, img_channels = 224, 224, 3  # Assuming RGB images
X, y = make_classification(n_samples=n_samples, n_features=img_height * img_width * img_channels,
                           n_classes=298, random_state=42)
X = X.reshape((n_samples, img_height, img_width, img_channels))  # Reshape to image format

# Normalize features (if needed)
X = X.astype('float32') / 255.0

# Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Convolutional Autoencoder
cae = ConvolutionalAutoencoder(input_shape=(img_height, img_width, img_channels), n_classes=len(np.unique(y)))
cae.train_autoencoder(X_train)

# Extract encoded features from the autoencoder for classification
encoder = Model(cae.autoencoder.input, cae.autoencoder.layers[-4].output)  # Get the output of the last encoder layer
X_train_encoded = encoder.predict(X_train)
X_test_encoded = encoder.predict(X_test)

# Train the classifier using the encoded features
cae.train_classifier(X_train_encoded, y_train)

# Evaluate the classifier
test_loss, test_accuracy = cae.classifier.evaluate(X_test_encoded, y_test, verbose=0)
print(f'Test Accuracy: {test_accuracy:.4f}')
```
