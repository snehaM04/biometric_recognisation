#  Loading the image dataset

```python
import os
import cv2
import numpy as np

def load_images_and_labels(base_dir, img_size=(224, 224)):
    images = []
    labels = []
    
    # Loop over each folder in the dataset directory
    for folder_name in os.listdir(base_dir):
        folder_path = os.path.join(base_dir, folder_name)
        
        # Ensure that we are only processing directories (i.e., label folders)
        if os.path.isdir(folder_path):
            # Loop over each image in the folder
            for image_name in os.listdir(folder_path):
                img_path = os.path.join(folder_path, image_name)
                print(f"Loading image: {img_path}")  # Debugging line
                
                # Read and resize the image
                img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                if img is None:
                    print(f"Failed to load image: {img_path}")
                    continue
                
                img = cv2.resize(img, img_size)  # Resize to match EfficientNet input size
                
                # Append the image and the label (folder name)
                images.append(img)
                labels.append(int(folder_name))  # Assuming folder name is the label (e.g., 1, 2, 3, etc.)
    
    # Convert lists to NumPy arrays
    images = np.array(images)
    labels = np.array(labels)
    
     #Normalize images to the range [0, 1]
    images = images.astype(np.float32) / 255.0

    print(f"Loaded {len(images)} images after quality check.")
    print("Images shape:", images.shape)
    print("Labels shape:", labels.shape)

    return images, labels

# Load images and labels
image_dir = "C:/iris_dataset"  # Replace with the correct path
images, labels = load_images_and_labels(image_dir)
```

```python
# Prepare images for EfficientNet (resize, add channels, normalize)
def prepare_images_for_efficientnet(images, target_size):
    # Resize, add channel, convert to 3-channel format for EfficientNet
    processed_images_resized = [cv2.resize(image, target_size) for image in images]
    processed_images_resized = np.stack(processed_images_resized, axis=0)
    processed_images_resized = np.expand_dims(processed_images_resized, axis=-1)  # Add channel dimension (grayscale)
    processed_images_resized = np.repeat(processed_images_resized, 3, axis=-1)    # Convert grayscale to 3 channels
    # Check for NaN or Inf values and clip them
    if np.isnan(processed_images_resized).any() or np.isinf(processed_images_resized).any():
        print("Warning: NaN or Inf found in the images! Fixing them...")
        processed_images_resized = np.nan_to_num(processed_images_resized, nan=0.0, posinf=255.0, neginf=0.0)
    
    # Normalize images using EfficientNet's preprocess_input
    processed_images_resized = processed_images_resized.astype(np.float32)
    processed_images_resized = preprocess_input(processed_images_resized)
    
    return processed_images_resized

# Extract features using EfficientNet model
def extract_features_with_efficientnet(images, model, target_size):
    processed_images = prepare_images_for_efficientnet(images, target_size)
    
    # Extract features using the EfficientNet model
    features = model.predict(processed_images)
    print(type(features))
    
    return features
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Input
from tensorflow.keras.models import Model
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.optimizers import Adam
from sklearn.datasets import make_classification

# OPA (Orca Predator Algorithm) for Feature Optimization
class OrcaPredatorAlgorithm:
    def __init__(self, n_predators, n_features, max_iter, features, labels):
        self.n_predators = n_predators
        self.n_features = n_features
        self.max_iter = max_iter
        self.features = features
        self.labels = labels
        self.position = np.random.randint(2, size=(n_predators, n_features)).astype(float)  # Convert to float after initialization
        self.velocity = np.random.uniform(-1, 1, size=(n_predators, n_features))  # Velocity
        self.best_position = np.copy(self.position)  # Best position
        self.best_fitness = np.full(n_predators, -np.inf)  # Best fitness for each predator
        self.global_best_position = np.zeros(n_features)  # Global best position
        self.global_best_fitness = -np.inf  # Global best fitness

    # Function to evaluate fitness based on feature selection
    def evaluate_fitness(self, predator_position):
        selected_features = np.where(predator_position == 1)[0]  # Get selected feature indices
        
        # If no features are selected, return a low fitness value
        if len(selected_features) == 0:
            return -np.inf

        # Use only the selected features
        X_selected = self.features[:, selected_features]
        
        # Split the data into train and test sets
        X_train, X_test, y_train, y_test = train_test_split(X_selected, self.labels, test_size=0.2, random_state=42)

        # Define a simple neural network model for classification
        model = self.build_classifier(X_train.shape[1])
        
        # Train the model
        model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=0)
        
        # Evaluate on the test set
        y_pred = model.predict(X_test)
        y_pred = (y_pred > 0.5).astype(int)  # Convert probabilities to binary predictions
        accuracy = accuracy_score(y_test, y_pred)
        
        return accuracy

    # Function to build a simple classifier model
    def build_classifier(self, input_shape):
        inputs = Input(shape=(input_shape,))
        x = Dense(64, activation='relu')(inputs)
        x = Dense(32, activation='relu')(x)
        output = Dense(1, activation='sigmoid')(x)
        
        model = Model(inputs=inputs, outputs=output)
        model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
        return model
    
    # Orca movement behavior
    def move_predators(self):
        for i in range(self.n_predators):
            # Update velocity based on best predator
            self.velocity[i] = 0.5 * self.velocity[i] + np.random.uniform(0, 1) * (self.best_position[i] - self.position[i]) + np.random.uniform(0, 1) * (self.global_best_position - self.position[i])
            
            # Update position
            self.position[i] += self.velocity[i]
            
            # Ensure the position remains binary (0 or 1)
            self.position[i] = np.clip(self.position[i], 0, 1)
        
    # Update best position and fitness
    def update_best_position(self):
        for i in range(self.n_predators):
            fitness = self.evaluate_fitness(self.position[i])
            if fitness > self.best_fitness[i]:
                self.best_fitness[i] = fitness
                self.best_position[i] = self.position[i]
            
            # Update global best
            if fitness > self.global_best_fitness:
                self.global_best_fitness = fitness
                self.global_best_position = self.position[i]
    
    # Run the OPA algorithm
    def run(self):
        for iteration in range(self.max_iter):
            print(f"Iteration: {iteration + 1}/{self.max_iter}")
            self.update_best_position()
            self.move_predators()

        return self.global_best_position, self.global_best_fitness


# Example feature extraction using EfficientNet
input_shape = (224, 224, 3)  # Assuming images are 224x224x3
num_images = 1010  # Number of images in the dataset

# Run OPA to optimize the feature set
n_predators = 2  # Number of predators (agents)
max_iter = 6 # Number of iterations

# Initialize and run the OPA algorithm
# opa = OrcaPredatorAlgorithm(n_predators=n_predators, n_features=features_flat.shape[1], 
#                             max_iter=max_iter, features=features_flat, labels=labels)

opa = OrcaPredatorAlgorithm(n_predators=n_predators, n_features=scaled_features.shape[1], 
                            max_iter=max_iter, features=scaled_features, labels=labels)


# Run the optimization
best_features, best_fitness = opa.run()

print(f"Best Fitness: {best_fitness}")
print(f"Best Feature Set: {best_features}")
```
